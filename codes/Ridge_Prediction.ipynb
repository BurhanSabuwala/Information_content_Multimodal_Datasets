```
import pandas as pd
from sklearn.decomposition import PCA 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
```

```
df_RNA = pd.read_csv("/content/drive/MyDrive/Curriculars/AACB_course_project/Datasets/RNA_features_scaledata_BMNC.csv", index_col = 0)
df_ADT = pd.read_csv("/content/drive/MyDrive/Curriculars/AACB_course_project/Datasets/ADT_features_scaledata_BMNC.csv", index_col = 0)
df_RNA.head()
```
